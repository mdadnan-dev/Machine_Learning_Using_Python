## Logistic Regression MultiClass Classification 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python
digits.target[0:10]
```

```python
from sklearn import model_selection

X_train, X_test, Y_train, Y_test = model_selection.train_test_split(digits.data,digits.target,train_size=0.8, test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn import linear_model
```

```python
logistic_model = linear_model.LogisticRegression(max_iter= 10000)
logistic_model.fit(X_train, Y_train)
```

```python
logistic_model.score(X_test,Y_test)
```

```python
plt.matshow(digits.images[98])
```

```python
digits.target[98]
```

```python
logistic_model.predict([digits.data[98]])
```

```python
# digits.image  --> this holds different images of digits from 0-9
# digits.data  -->  this holds data of the digits in an array format
# digits.target --> this holds the actual value of the digit 
```

```python
predicted = logistic_model.predict(X_test)
```

```python
from sklearn import metrics
cm = metrics.confusion_matrix(Y_test,predicted)
```

```python
import seaborn as sns
plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True)

# To give a basic understanding of the confusion matrix
# Y axis - 7 and X axis - 5   The box shows value 2 
# This means the digit was 7 but the model said its 5 , 2 times 

```
