## Logistic Regression Excercise 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Datasets/HR_comma_sep.csv")
data.head()
```

```python
data.groupby('left').mean(numeric_only=True)

# So from this we can select columns that effect a person from leaving the company 
# Satisfaction level 
# average_montly_hours	
# promotion_last_5years
# Salary
```

```python
data_dummies = pd.get_dummies(data['salary'])
```

```python
data_dummies = data_dummies[['high','low','medium']].replace({True: 1, False: 0})
data_dummies
```

```python
data = pd.concat([data, data_dummies], axis='columns')
```

```python
data
```

```python
data = data.drop(columns= ['salary'])
```

```python
data.head()
```

```python
from sklearn import model_selection
```

```python
X = data[['satisfaction_level', 'average_montly_hours','promotion_last_5years','high','low','medium']]
Y = data[['left']]
```

```python
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X, Y, train_size=0.8, test_size= 0.2)
```

```python
from sklearn import linear_model
```

```python
logistic_model = linear_model.LogisticRegression()
logistic_model.fit(X, Y)
```

```python
logistic_model.predict(X_test)
```

```python
logistic_model.score(X_test, Y_test) # The model is 77% accurate
```
