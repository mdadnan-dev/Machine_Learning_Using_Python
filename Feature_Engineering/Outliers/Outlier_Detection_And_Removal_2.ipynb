```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv(r"Datasets/bhp.csv")
```

```python
data.describe()
```

```python
High_Price = data.price_per_sqft.quantile(0.99)
High_Price
```

```python
Low_price = data.price_per_sqft.quantile(0.05)
Low_price
```

```python
data[data['price_per_sqft']>High_Price]
```

```python
data[data['price']<Low_price]
```

```python
data= data[(data['price_per_sqft']<High_Price) & (data['price_per_sqft']>Low_price)]
```
