## Linear Regression on Dummies Excercise

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn import preprocessing
```

```python
data = pd.read_csv('Datasets/carprices.csv')
data
```

```python
x = data['Mileage']
y = data['Sell Price($)']
```

```python
plt.scatter(x,y, marker='^')
plt.show()
```

```python
data
```

```python
dummies = pd.get_dummies(data['Car Model'])
dummies
```

```python
dummies = dummies[['Audi A5','BMW X5', 'Mercedez Benz C class']].replace({True: 1, False:0})
```

```python
dummies
```

```python
data = pd.concat([data, dummies], axis='columns')
```

```python
data
```

```python
X = data[['Mileage', 'Age(yrs)','Audi A5','BMW X5','Mercedez Benz C class']]
Y = data['Sell Price($)']
```

```python
Reg_model = linear_model.LinearRegression()
Reg_model.fit(X,Y)
```

```python
Reg_model.predict([[4500, 4, 0, 0, 1]])
```

```python
Reg_model.predict([[8600, 7, 0, 1, 0]])
```

```python
Reg_model.score(X,Y)
```
