Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
hours_spent = np.array([20, 33, 54, 10, 90, 87, 59, 76, 43, 15]).reshape(-1, 1)

scores = np.array([65, 50, 60, 15, 100, 92, 78, 89, 77, 10]).reshape(-1, 1)
```

```python
print(hours_spent)
```

```python
print(scores)
```

```python
model = LinearRegression()
model.fit(hours_spent, scores)
```

```python
plt.scatter(hours_spent, scores)

plt.title('Linear Regression')
plt.xlabel('Hours Spent')
plt.ylabel('Scores Achieved')

plt.plot(np.linspace(0,100).reshape(-1,1), model.predict(np.linspace(0,100).reshape(-1, 1)), color = 'k')

plt.show()
```

```python
print(model.predict(np.array([76]).reshape(-1, 1)))  # Predicts if a person studies 76 hours He/She may score 88.03%
```

```python
print(model.predict(np.array([1]).reshape(-1,1))) # Predicts if a person studies 1 hour He/She  may score 19.51%
```
